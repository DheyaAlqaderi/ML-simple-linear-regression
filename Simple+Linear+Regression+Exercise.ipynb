## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

## Load the data

```python
data = pd.read_csv('real_estate_price_size.csv')
```

```python
data.head()
```

```python
data.describe()
```

## Create the regression

### Declare the dependent and the independent variables

```python
y = data['price']
x1 = data['size']
```

### Explore the data

```python
import matplotlib.pyplot as plt

# # Your data (x1 and y)
# x1 = [1, 2, 3, 4, 5]
# y = [10, 20, 25, 30, 35]

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(8, 6))

# Scatter plot with enhanced customizations
scatter = ax.scatter(x1, y, c='blue', marker='o', s=100, label='Data Points')

# Customize the axis labels and font size
ax.set_xlabel('Size', fontsize=20)
ax.set_ylabel('Price', fontsize=20)

# Set the title and font size (if needed)
ax.set_title('Scatter Plot', fontsize=24)

# Customize the tick label font size
ax.tick_params(axis='both', labelsize=16)

# Add a grid (optional)
ax.grid(True, linestyle='--', alpha=0.7)

# Add a legend (if multiple data sets are plotted)
ax.legend()

plt.xlim(0)
plt.ylim(0)

# Display the plot
plt.show()
```

### Regression itself

```python
# Add a constant term to the independent variable
x = sm.add_constant(x1)

# Fit the OLS regression model
results = sm.OLS(y, x).fit()

# Display the summary with enhanced formatting
from IPython.display import HTML, display

# Convert the summary table to HTML format
summary_html = results.summary().tables[1].as_html()

# Create an HTML object
summary_html_obj = HTML(summary_html)

# Display the HTML object
display(summary_html_obj)
```

### Plot the regression line on the initial scatter

```python
plt.scatter(x1,y)
yhat = 223.1787*x1 + 1.019e+05
# print(yhat)
fig = plt.plot(x1,yhat, lw=4, c='orange', label='regression line')
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.xlim(0)
plt.ylim(0)
plt.show()
```
